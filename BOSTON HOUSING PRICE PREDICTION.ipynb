```python
def outlier_removal(df1,column):
    
    Q1  = df1[column].quantile(0.25)
    Q3 = df1[column].quantile(0.75)
    IQR = Q3-Q1
    lower_bound = Q1 - IQR*1.5
    upper_bound = Q3 + IQR*1.5
    df1[column] = np.where(df1[column] < lower_bound, lower_bound, df1[column])
    df1[column] = np.where(df1[column] > upper_bound, upper_bound, df1[column])
    return df1
for column in df1.columns:
    df1 = outlier_removal(df1,column)
    
```

```python
sns.boxplot(df1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
x=df1.drop('medv',axis=1)
```

```python
y=df1.medv
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model1=LinearRegression()
model1.fit(x_train,y_train)
```

```python
y_pred=model1.predict(x_test)
```

```python
y_pred
```

```python
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
```

```python
print(f"mean squared error:{mse}")
print(f"r squared score:{r2}")
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x=y_test,y=y_pred)
plt.xlabel("actual price")
plt.ylabel("predicted price")
plt.title("Actual vs Predicted price")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=y_test,y=y_pred)
plt.xlabel("actual price")
plt.ylabel("predicted price")
plt.title("Actual vs Predicted price")
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
rf_model=RandomForestRegressor(n_estimators=100,random_state=42)
rf_model.fit(x_train,y_train)
```

```python
rf_pred=rf_model.predict(x_test)
rf_pred
```

```python
mse=mean_squared_error(y_test,rf_pred)
r2=r2_score(y_test,rf_pred)
print(f"mean squared error:{mse}")
print(f"r squared score:{r2}")
```

```python
dt_model=DecisionTreeRegressor(random_state=42)
dt_model.fit(x_train,y_train)
```

```python
dt_pred=dt_model.predict(x_test)
dt_pred
```

```python
mse=mean_squared_error(y_test,dt_pred)
r2=r2_score(y_test,dt_pred)
print(f"mean squared error:{mse}")
print(f"r squared score:{r2}")
```

```python
plt.figure(figsize=(20,15))
sns.lineplot(x=y_test,y=y_pred,label="linear regresson",color='blue')
sns.lineplot(x=y_test,y=dt_pred,label="decision tree",color='red')
sns.lineplot(x=y_test,y=rf_pred,label="random forest",color='yellow')
plt.xlabel("actual price")
plt.ylabel("predicted price")
plt.title("Actual vs Predicted price obtained from linear regression, decision tree, random forest")
plt.show()
```
